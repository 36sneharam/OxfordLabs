```
import numpy as np
import cv2
from matplotlib import pyplot as plt
```

```
filel = "MarsLandL.jpg"
filer = "MarsLandR.jpg"
```

```
from PIL import Image
imgl = Image.open(filel)
imgr = Image.open(filer)
Nimgl = imgl.resize((500,782))   # image resizing
Nimgr = imgr.resize((500,782))

Nimgl.save('Nimgl.jpg')
Nimgr.save('Nimgr.jpg')
```

```
plt.imshow(Nimgl)
```

```
plt.imshow(Nimgr)
```

```
imgL = cv2.imread('Nimgl.jpg',0)
imgR = cv2.imread('Nimgr.jpg',0)
stereo = cv2.StereoBM_create(numDisparities=64, blockSize=15)
disparity = stereo.compute(imgL,imgR)
plt.imshow(disparity,'gray')
plt.show()
```
